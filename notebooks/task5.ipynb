```python
from algorithms_sc.recursive.queens import solve_queens

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from tqdm.auto import tqdm
```

```python
MAX_N = 29
ns = range(4, MAX_N + 1)
```

```python
counts = []
arrangements = []
for n in tqdm(ns):
    current_desk = solve_queens(n)
    if current_desk == -1:
        raise ValueError
    else:
        counts.append(current_desk.try_count)
        arrangements.append(current_desk.desk)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(16, 9))
ax.plot(ns, counts)
ax.set_yscale('log')
plt.show()
```
